# 反向传播

```python
import torch
import matplotlib.pyplot as plt
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
# Tensor张量有两个值，权重本身的data和损失对权重的梯度grad，注意grad也是Tensor张量，grad初始为None
w = torch.Tensor([1.0])
# 这句话的意思是它是需要计算梯度的
w.requires_grad = True
```

```python
# 模型
def forward(x):
    # 这个x自动类型转换为Tensor    
    return x * w
# 损失函数
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y)*(y_pred - y)
```

```python
epoch_list = []
l_list = []
```

```python
print("predict(before training)",4,forward(4).item())
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        # 前馈,计算图        
        l = loss(x,y)
        # 反馈        
        # 梯度会自动存储到w里的grad，backward后计算图就释放了        
        l.backward()
        # item()返回一个标量,该方法的功能是以标准的Python数字的形式来返回这个张量的值        
        print('\tgrad:',x,y,w.grad.item())
        # 调用l.backward()方法后w.grad为Tensor，故更新w.data要使用w.grad.data，w.grad.item()应该也行吧？        
        w.data = w.data-0.01*w.grad.data
        # 求得的梯度要清零，防止影响下一次的梯度        
        w.grad.data.zero_()
    epoch_list.append(epoch)
    l_list.append(l.item())
    print("progress:",epoch,l.item())
print("predict(after training)",4,forward(4).item())
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.plot(epoch_list,l_list)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
